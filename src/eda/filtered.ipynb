```python
import pandas as pd
```

```python
# Đường dẫn tới file CSV
file_path = r"D:\HUST\_Intro to DS\Capstone Project\Football-Match-Prediction\src\eda\eda_data\selected\all_feature_names.csv"

# Đọc file CSV chứa danh sách tên cột
feature_names_df = pd.read_csv(file_path)

df =pd.read_csv('../../data/processed/la_liga_data.csv')
```

```python
# Đọc file CSV chứa danh sách tên cột


# Lấy danh sách các tên cột từ cột Feature_Name
feature_names = feature_names_df['Feature_Name'].dropna().unique()

# Lọc DataFrame gốc để chỉ giữ các cột có tên nằm trong feature_names
df_filtered = df[[col for col in feature_names if col in df.columns]]

# In DataFrame đã lọc
df_filtered.head()
```

# Manual remove cols 

```python
# Xóa các cột 'Performance__GA' và 'Penalty Kicks__PKatt' từ df_filtered
df_filtered = df_filtered.drop(columns=['Performance__GA', 'Penalty Kicks__PKatt'], errors='ignore')

# In kết quả DataFrame sau khi chỉnh sửa
df_filtered.head()
```

```python
df_filtered.to_csv('filtered_df.csv')
```
