```python
import pandas as pd
```

```python
# Đường dẫn tới file CSV
file_path = "eda_data/selected/all_feature_names_rest.csv"

# Đọc file CSV chứa danh sách tên cột
feature_names_df = pd.read_csv(file_path)

df =pd.read_csv('../../data/processed/la_liga_data.csv')
df_pre_filtered = pd.read_csv('filtered_df.csv')
```

```python
# Đọc file CSV chứa danh sách tên cột


# Lấy danh sách các tên cột từ cột Feature_Name
feature_names = feature_names_df['Feature_Name'].dropna().unique()

# Lọc DataFrame gốc để chỉ giữ các cột có tên nằm trong feature_names
df_filtered = df[[col for col in df.columns if col in feature_names or col in df_pre_filtered.columns]]

# In DataFrame đã lọc
df_filtered.head()
```

```python
df_filtered.to_csv('filter__rest_df.csv')
```
